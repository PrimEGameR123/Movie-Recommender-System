now i want only three columns, 1. id, 2. movie name & 3. tags.<br>
Tags will contain all the data i.e., genres, cast and crew.

```python
# for understanding only
import ast
print(ast.literal_eval(df.iloc[0].genres))
print(type(df.iloc[0].genres))
print(type(ast.literal_eval(df.iloc[0].genres)))
```

```python
# we need to convert above dictionary series into an array, this is basically preprocessing. so creating a function for that.
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres'].apply(convert)
```

```python
df['genres']=df['genres'].apply(convert)
df
```

```python
df.iloc[0].keywords
```

```python
# similarly
df['keywords']=df['keywords'].apply(convert)
```

```python
df.iloc[0].cast
```

```python
df.iloc[0].crew
```

```python
# as cast is very huge, we will take only first 10 names without duplicates
def convert10(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=10:
            if i['name'] not in L:
                L.append(i['name'])
            else:
                continue
            counter+=1
    return L
```

```python
df['cast']=df['cast'].apply(convert10)
```

```python
df
```

```python
df.iloc[0].crew
```

```python
# we want only that names from crew which are producer and director
def fetch_dir_pro(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            if i['name'] not in L:
                L.append(i['name'])
        if i['job']=='Producer':
            if i['name'] not in L:
                L.append(i['name'])
    return L
```

```python
df['crew']=df['crew'].apply(fetch_dir_pro)
```

```python
df
```

as it is seen that overview column is a string, so converting it into list as well so that we can concatenate it with other columns easily.

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df
```

```python
#removing stopping words like 'and', 'is', 'the' etc. from overview because these are not our keywords.
from nltk.corpus import stopwords
```

```python
print(stopwords.words("english"))
```

```python
for i in np.arange(1,df['overview'].shape[0]):
    word_without_stopwords=[words for words in df['overview'].iloc[i] if words not in stopwords.words("english")]
    df['overview'].iloc[i]=word_without_stopwords
```

Now we need to remove the spaces between the keywords, e.g., 
Sam Worthington & Sam Mendes may confuse the algorithm if we use 'Sam' for prediction.

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df
```

```python
# making a new column tag which contain all column from overview to crew
df['tags']=df['genres']+df['keywords']+df['cast']+df['crew']+df['overview']
```

```python
new_df=df[['id','title','tags']]
```

```python
new_df.reset_index(drop=True,inplace=True)
```

```python
new_df
```

```python
# doing word stemming for getting the same root word e.g., [Loving,Love,Loved] to [Love,Love,Love].
import nltk
from nltk.stem.snowball import SnowballStemmer
stemmer=SnowballStemmer(language="english")
```

```python
def stemWords(words):
    y=[]
    for i in words:
        y.append(stemmer.stem(i))
    return y
```

```python
stemWords(df.iloc[0].tags)
```

```python
new_df['tags']=new_df['tags'].apply(stemWords)
```

```python
new_df
```

```python
temp_df=new_df.copy()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
# vectorising the tags columns and taking 8000 most occuring words from tags for vectorising by using method 
## called Bag of words (in this, a text is represented as the bag of its words), then making the array of all movies
### as per the tags in them and vectors assigned to them, this will make a table of (4800,8000) of vectorizers.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=8000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
print(vectors)
```

```python
np.count_nonzero(cv.get_feature_names())
```

```python
cv.get_feature_names()
```

```python
# using cosine similarity for checking vectors similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
# that means a matrix of 4800, 4800 will be made and each movie name similairty percentage will be checked for all movies.
## 1 means most similar and 0 means most dissimilar.
### similarity is checked on the basis of not eucledian distance but angle between them, because in high dimensions, eucledian distance fails.
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
# sample
# now creating a function to take out the top 5 similar movies
sorted(enumerate(similarity[5]),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    similar=sorted(enumerate(similarity[movie_index]),reverse=True,key=lambda x:x[1])[1:8]
    for i in similar:
        print(new_df['title'][i[0]])
```

```python
recommend('Batman Begins')
```
